## Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import datetime as dt
```

## Reading the data

```python
df = pd.read_csv("netflix1.csv")
```

```python
df.head(10)
```

```python
df.drop_duplicates(subset ="show_id",
                     keep = False, inplace = True)
df.head()
```

```python
df.info()
```

## Creating new columns

```python
df["date_added"] = pd.to_datetime(df["date_added"])
df.info()
```

```python
df['month_added'] = df['date_added'].dt.month_name()
df
```

```python
df['year_added'] = df['date_added'].dt.year
df
```

```python
df['day_added']=df['date_added'].dt.day_name()
df
```

## Types of Shows on Netflix

```python
types =df.groupby(['type',])[ 'type'].count().reset_index(name='count')
types=types.set_index('type')
types
```

```python
colors = ['#ff9999','#66b3ff']
types.plot.pie(y='count',autopct='%.1f%%', shadow=True, legend= 'type' , figsize=(6,6),colors=colors)
plt.title('types_of_show', fontsize=20)
```

## Group By Directors of show and Type of show

```python
show_director= df.groupby(['director','type'])['type'].count().reset_index(name = 'show_count')
```

```python
show_director
```

```python
top10_directors=show_director[show_director['show_count'] >= 12 ]
top10_directors.sort_values(by = ['show_count'], ascending = False)
```

## Group By Country of Show and type of Show

```python
show_origin= df.groupby(['country','type'])[ 'type'].count().reset_index(name = 'show_count')
show_origin
```

## Top 10 Countries to produce movies

```python
show_origin_movie  = show_origin[show_origin['type'] == 'Movie']
top10 = show_origin_movie.sort_values(by = ['show_count'], ascending = False)[0:10]
top10
```

```python
sb.barplot(x =top10.reset_index()['country'], y= top10.reset_index()['show_count']);
plt.title('Top 10 Countries to produce Movies', fontsize=16)
plt.xlabel('Country', fontsize=16);
plt.ylabel('Movies produced', fontsize=16);
sb.set(rc = {'figure.figsize':(20,10)})
plt.xticks(rotation = 45, fontsize=16)
plt.yticks(fontsize=16)
```

```python
show_origin_tv  = show_origin[show_origin['type'] == 'TV Show']
top10_show = show_origin_tv.sort_values(by = ['show_count'], ascending = False)[0:10]
top10_show
```

## Top 10 countries to produce TV Shows

```python
sb.barplot(x =top10_show.reset_index()['country'], y= top10_show.reset_index()['show_count']);
plt.title('Top 10 Countries to produce TV Shows', fontsize=16)
plt.xlabel('Country', fontsize=16);
plt.ylabel('TV Shows produced', fontsize=16);
sb.set(rc = {'figure.figsize':(20,10)})
plt.xticks(rotation = 45, fontsize=16)
plt.yticks(fontsize=16)
```

## Group By Rating and Show Type

```python
show_rating= df.groupby(['rating','type'])['rating'].count().reset_index(name='ratings_count')
show_rating=show_rating.set_index('rating')
show_rating
```

```python
show_rating_movie = show_rating[show_rating['type'] == 'Movie']
show_rating_movie
```

## Distribution of Movie Ratings

```python
explode = (0.05, 0.05, 0.05, 0.05, 0.05, 0.05, 0.05, 0.05, 0.05, 0.05, 0.05, 0.05, 0.05, 0.05 )
show_rating_movie.plot.pie(y='ratings_count',autopct='%.1f%%', figsize=(13,17), pctdistance= 0.80,  explode=explode)
plt.title('Movie ratings distribution', fontsize=20)
plt.ylabel(" ")
centre_circle = plt.Circle((0, 0), 0.82, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.legend(ncol=15, loc="upper center")
```

```python
show_rating_tv = show_rating[show_rating['type'] == 'TV Show']
show_rating_tv
```

## Distribution of TV Show ratings

```python
explode = (0.05, 0.05, 0.05, 0.05, 0.05, 0.05, 0.05, 0.05, 0.05,)
show_rating_tv.plot.pie(y='ratings_count',autopct='%.1f%%', figsize=(13,17), pctdistance= 0.80)
plt.title('Distribution of TV Show ratings', fontsize=20)
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.legend(ncol=15, loc="upper center")
```

## Grouped by Day added and Show type

```python
Day_added= df.groupby(['day_added','type'])[ 'type'].count().reset_index(name='Day_count')
Day_added
```

```python
Day_added_movie = Day_added[Day_added['type'] == 'Movie']
Day_added_movie.sort_values(by = ['Day_count'], ascending = True, inplace = True)
Day_added_movie 
```

## Movies released on each day

```python
sb.barplot(x =Day_added_movie.reset_index()['day_added'], y= Day_added_movie.reset_index()['Day_count']);
plt.title('No of Movies Released each day of the week in Ascending order' , fontsize=20)
plt.xlabel('Day of week', fontsize=16);
plt.ylabel('Movies count', fontsize=16);
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
sb.set(rc = {'figure.figsize':(24,12)})
```

```python
Day_added_tv = Day_added[Day_added['type'] == 'TV Show']
Day_added_tv.sort_values(by = ['Day_count'],ascending = True, inplace = True)
Day_added_tv                        
```

## TV Shows released on each day

```python
sb.barplot(x =Day_added_tv.reset_index()['day_added'], y= Day_added_tv.reset_index()['Day_count']);
plt.title('No of Shows Released each day of the week in Ascending order' , fontsize=20)
plt.xlabel('Day of week', fontsize=16);
plt.ylabel('Show count', fontsize=16);
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
sb.set(rc = {'figure.figsize':(24,12)})
```

## Movies and TV Shows released each day of the week

```python
sb.barplot(x =Day_added.reset_index()['day_added'], y= Day_added.reset_index()['Day_count'],
            hue =  Day_added.reset_index()['type']);
plt.title('No of Movies and Shows Released each day of the week' , fontsize=16)
plt.xlabel('Day of week', fontsize=16);
plt.ylabel('Show count', fontsize=16);
sb.set(rc = {'figure.figsize':(24,12)})
```
